```python
#Tahap Preprocessing
file = open('abstracts.wiki.txt','r', encoding= 'utf8')
data = []
for w in file.readlines():
    data.append(w)
len(data)

dataForPreprocessing = data[:1000]
print(dataForPreprocessing)
    
```

```python
# Tokenization and remove punctuation with NLTK

import nltk
import string
import re
from collections import Counter
from string import digits

tokenize= []
def tokenization_removePunctuation(kalimat, tokens):
    #remove number
    kalimat = ''.join([i for i in kalimat if not i.isdigit()])
    kalimat = re.sub('r ^ https ?:\/\/.*[\r\n]*','', kalimat, flags=re.MULTILINE)
    translator = str.maketrans('','', string.punctuation)
    #remove punctuation
    no_punctuation = kalimat.translate(translator)
    for token in nltk.word_tokenize(no_punctuation):
        tokens.append(token)
        count = Counter(tokens)
        
for kalimat in dataForPreprocessing:
    tokenization_removePunctuation(kalimat, tokenize)
```

```python
#Case Folding
def caseFolding(tokens):
    tokens = [kata.lower() for kata in tokens]
    return tokens
tokenize = caseFolding(tokenize)

```

```python
token = tokenize
tokenize = []
def decodeAsciiCode(kalimat): return "".join(i for i in kalimat if ord(i)<128)
for words in token:
    tokenize.append(decodeAsciiCode(words))
all_sentence =' '.join(tokenize)    
```

```python
#Stopping (Stop words) with NLTK
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

#Stopwords = tokenize
filtered = [w for w in tokenize if not w in stopwords.words('english')]
count = Counter(filtered)
print(filtered)
```

```python
#Normalization (Stemmer) with NLTK (Porter Stemmer)
stemmer = PorterStemmer()
def stemming(tokens):
    for i in range(0, len(tokens)):
        if(tokens[i] !=stemmer.stem(tokens[i])):
            tokens[i] = stemmer.stem(tokens[i])
        return tokens

tokenize = stemming(filtered)
```

```python
from operator import itemgetter
import matplotlib.pyplot as plot
import numpy as np
import re
import pandas as pd
from pandas import DataFrame
```

```python
frequency = {}
for word in tokenize:
    count = frequency.get(word,0)
    frequency[word] = count + 1

first_digit_frequency = []
for key, value in reversed(sorted(frequency.items(), key = itemgetter(1))):
    first_digit = str(value)[:1]
    first_digit_frequency.append(first_digit)
    print (key, value)
    
data_new = []
for key, value in reversed(sorted(frequency.items(), key=itemgetter(1))):
    data_new.append({'Terms':key, 'Freq': value})
```

```python
df = pd.DataFrame(data_new, columns=["Terms","Freq"])
i=1
temp_rank = []
temp_data = df.sort_values(by=['Freq'], ascending=False)
temp_data

for w in temp_data['Freq']:
    temp_rank.append(i)
    i=i+1
    
temp_data['Rank'] = temp_rank
temp_data['r x freq'] = temp_data['Rank'] * temp_data['Freq']
temp_data

df_sorted = temp_data.sort_values(by=['Rank'], ascending=True)
df_sorted
```

```python
%matplotlib inline
data = df_sorted.head(10).plot.bar()
data.set_xticklabels(labels = temp_data['Terms'])
```

```python
from math import log10
from matplotlib import pyplot as plt

def plot_benford(iterable, title):
    numbers = [float(n) for n in range(1,10)]
    benford = [log10(1+1/d) for d in numbers]
    plot.plot(numbers, benford, 'ro--', label="Distribusi Hukum Benford")
    plot.plot(numbers, iterable, 'bs--', label="Distribusi Wikipedia")
    
    plot.title(title)
    plot.xlabel("Digit")
    plot.ylabel("Freq")
    
    plot.xlim(0.10)
    plot.xticks(numbers)
    plot.legend()
    plot.show()
    
def digits(iterable):
    numexp = re.compile(r'\d+(\.\d+)?([eE]\d+)?')
    leading = set("123456789")
    
    for item in iterable:
        for match in numexp.finditer(str(item)):
            for digit in match.group(0):
                if digit in leading:
                    yield int(digit)
                    break

data = df_sorted.head(9)
fd = [log10(1+(1/float(str(n)[0])))for n in data['r x freq']]
plot_benford(fd,"Hukum Benford menggunakan digit pertama dalam r x freq")

fd_abai = [log10(1+(1.0/(n)))for n in data['r x freq']]
plot_benford(fd_abai,"Hukum Benford dengan mengabaikan Digit Pertama dari r x freq")
```
